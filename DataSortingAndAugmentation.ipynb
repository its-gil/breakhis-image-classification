```python
# get the Labels of the filepath Substrings into own Collums
imageDataframe["TumorClass"]=imageDataframe["filename"].str.split("_|-",regex=True).str[1]
imageDataframe["Subclass"]=imageDataframe["filename"].str.split("_|-",regex=True).str[2]
imageDataframe
```

```python
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report
```

```python
classEncoder = LabelEncoder()
imageDataframe['TumorClass'] = classEncoder.fit_transform(imageDataframe['TumorClass'])
imageDataframe
subclassencoder= OneHotEncoder(sparse_output=False) # Set sparse=False for a dense output
subclass_encoded = subclassencoder.fit_transform(imageDataframe[['Subclass']])
encoded_columns = subclassencoder.get_feature_names_out(['Subclass'])
imageDataframe[encoded_columns] = subclass_encoded
imageDataframe = imageDataframe.drop(columns=['Subclass'])
imageDataframe
```

```python
train_images_df, val_images_df = train_test_split(imageDataframe, test_size=0.2, random_state=42, stratify=imageDataframe['TumorClass'])
```

```python
from PIL import Image
import numpy as np
targetfolder="/content/drive/MyDrive/ProjectDataPreprzessing/lowRes255/"
```

```python
def resize_image(trainTest,img_path, target_size=(256, 256)):
    picturePath=f"/content/drive/MyDrive/ProjectDataPreprzessing/processedData/ {img_path}"
    img = Image.open(picturePath)
    img = img.resize(target_size)
    img.save(f"{targetfolder}{img_path}")
```

```python
# scale Images
for index, row in imageDataframe.iterrows():
  print(index)
  resize_image(row["split"],row["filename"])
```

```python
train_images_vector=[]
train_labels_binary=[]
train_labels_multclass=[]
i=0
for index, row in train_images_df.iterrows():
  i+=1
  img = Image.open(f"{targetfolder}{row['filename']}")
  img_array = np.array(img)
  img_array=img_array/255.0
  train_images_vector.append(img_array)
  # the binary class label vector is is we have Tumor Class B or M (B yes or no)
  train_labels_binary.append(row["TumorClass"])
  #th multclass label vector is the one part of the Row that reresents the one hot encoded Sublabels
  #train_labels_multclass.append(row[["Subclass_A","Subclass_DC","Subclass_F","Subclass_LC","Subclass_PT","Subclass_TA"]])
  print(f"row {index} -> {i}")
  if i>2000:
    break
```

```python
val_images_vector=[]
val_labels_binary=[]
val_labels_multclass=[]
y=0
for index, row in val_images_df.iterrows():
  y+=1
  img = Image.open(f"{targetfolder}{row['filename']}")
  img_array = np.array(img)
  img_array=img_array/255.0
  val_images_vector.append(img_array)
  val_labels_binary.append(row["TumorClass"])
  if y>200:
    break
  #val_labels_multclass.append(row[["Subclass_A","Subclass_DC","Subclass_F","Subclass_LC","Subclass_PT","Subclass_TA"]])
```

```python
#Create Model.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model,Sequential
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
```

```python
model = Sequential()
model.add(Conv2D(32, (4, 4), activation='relu', input_shape=(256, 256, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (4, 4), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(np.array(train_images_vector), np.array(train_labels_binary), epochs=10, batch_size=32, validation_data=(np.array(val_images_vector), np.array(val_labels_binary)))
```
